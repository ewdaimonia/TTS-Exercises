### Exercise 1. 

Read Total profit of all months and show it using a line plot

Total profit data provided for each month. Generated line plot must include the following properties:

1. X label name = Month Number
2. Y label name = Total profit

````python
# Exercise 1
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
import pandas as pd

company_sales = pd.read_csv("data/company_sales_data.csv")
# company_sales.describe()
# company_sales

fig = plt.figure()
ax = plt.axes()
plt.plot(company_sales.month_number, company_sales.total_profit)
ax.set(xlim=(1, 12), ylim=(150000, 450000),
       xlabel='Month Number', ylabel='Total profit',
       title='Exercise 1');
````

### Exercise 2. 

Get Total profit of all months and show line plot with the following Style properties

Generated line plot must include following Style properties:

1. Line Style dotted and Line-color should be red
2. Show legend at the lower right location.
3. X label name = Month Number
4. Y label name = Sold units number
5. Add a circle marker.
6. Line marker color as read
7. Line width should be 3

````python
# Exercise 2
ax = plt.axes(label="2")
plt.plot(company_sales.month_number, company_sales.total_profit, linestyle='dotted', color='red', label='Total Profit Over Time', marker="o", linewidth=3, markerfacecolor='red')
# plt.plot(4, 350000, "o")
plt.legend(loc='lower right')
ax.set(xlim=(1, 12), ylim=(150000, 450000),
       xlabel='Month Number', ylabel='Sold units number',
       title='Exercise 2');
````

### Exercise 3.

Read all product sales data and show it using a multiline plot

Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).

````python
# Exercise 3facecream 	facewash 	toothpaste 	bathingsoap 	shampoo 	moisturizer 	
ax = plt.axes(label="3")
plt.plot(company_sales.month_number, company_sales.facecream, linestyle='dotted', color='red', label='Facecream')
plt.plot(company_sales.month_number, company_sales.facewash, linestyle='dashed', color='blue', label='Facewash')
plt.plot(company_sales.month_number, company_sales.toothpaste, linestyle='solid', color='cyan', label='Toothpaste')
plt.plot(company_sales.month_number, company_sales.bathingsoap, linestyle='dashdot', color='green', label='Bathing Soap')
plt.plot(company_sales.month_number, company_sales.shampoo, linestyle='solid', marker="o", color='black', label='Shampoo')
plt.plot(company_sales.month_number, company_sales.moisturizer, linestyle='dotted', marker="v", color='orange', label='Moisturizer')
plt.legend(loc='upper left')
ax.set(xlim=(1, 12), ylim=(0, 15000),
       xlabel='Month Number', ylabel='Sold units number',
       title='Exercise 3');
````

### Exercise 4.

Read toothpaste sales data of each month and show it using a scatter plot

Also, add a grid in the plot. gridline style should .

````python
# Exercise 4
ax = plt.axes(label="4")
plt.scatter(company_sales.month_number, company_sales.toothpaste, marker='o')
ax.grid(linestyle='-', linewidth='0.5', color='black')
ax.set(xlim=(1, 12), ylim=(4000, 10000),
       xlabel='Month Number', ylabel='Toothpaste sold units number',
       title='Exercise 4');
````

### Exercise 5.

Read face cream and face wash product sales data and show it using the bar chart

The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

````python
# Exercise 5
ax = plt.axes(label="5")
plt.bar(company_sales.month_number, company_sales.facewash, color='green', label='Facewash', width = 0.25)
plt.bar(company_sales.month_number + .25, company_sales.facecream, color='red', label='Facecream', width = 0.25)
plt.legend(loc='upper left')
ax.set(xlabel='Month Number', ylabel='Number of Units Sold',
       title='Exercise 5');
````
